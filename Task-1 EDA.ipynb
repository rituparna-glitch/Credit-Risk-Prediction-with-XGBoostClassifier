```python
def kdePlot(var):
    fig = plt.figure(figsize=(8,6))
    ax=sns.kdeplot(df.loc[(df.high_risk_applicant == 0),var] , color='g',shade=True, label='not high risk applicants') 
    ax=sns.kdeplot(df.loc[(df.high_risk_applicant == 1),var] , color='b',shade=True, label='high risk applicants')
    plt.legend()
    plt.title('High risk applicants with respect to {}'.format(var))

columns = ['Primary_applicant_age_in_years', 'Number_of_dependents', 'Years_at_current_residence', 'Months_loan_taken_for', 
           'Principal_loan_amount', 'EMI_rate_in_percentage_of_disposable_income', 'Number_of_existing_loans_at_this_bank']
for n in columns:
    kdePlot(n)
```

The density distribution of high risk and low risk applicants is almost similar for these continuous variables. 
The rough inferences we can draw are:
- For applicant age around 20-40 there are more high risk applicants 
- Applications with number of dependents 1 shows more high rish applicants. 
- Candidates with 2 and 4 years at current residence have more high risk applications.  
- Loan taken for 0-20 months show more high risk applicants. Whereas loan taken for 40-60 months includes more low risk applicants. 
- In between principal loan amount of 1000000-5000000 there are more high risk candidates. 
- 3-5% of EMI rate shows a greater number of high risk candidates. EMI rate around 2% shows more low risk candidates. 
- Candidates having around 1 and 2 number of existing loans at the bank are high risk. It shows the similar distribution for low risk candidates too.

## Skewness Completeness

Skewness essentially measures the symmetry of the distribution, while kurtosis determines the heaviness of the distribution tails.
- Kurtosis is the sharpness of the peak of a frequency-distribution curve .It is actually the measure of outliers present in the distribution.



```python
df.select_dtypes(include=['int64']).columns
```

```python
df.skew(axis = 0, numeric_only = None, skipna = False)
```

- Skewness between -0.5 to 0.5 means data is fairly symmetrical 
- skewness greater than 1 then data is highly skewed. 

- Primary_applicant_age_in_years, Months_loan_taken_for and Principal_loan_amount columns are highly right skewed.


```python
df.kurtosis(axis = 0)
```

- Positive kurtosis means data is pointy to the normal distribution. 
- Negative kurtosis means data is flatter than normal distribution
